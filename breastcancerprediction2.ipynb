Importing Libraries

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
from tensorflow.keras.layers import Convolution2D
from tensorflow.keras.layers import MaxPooling2D
from tensorflow.keras.layers import Flatten
```

Import Image Data Generator Library

```python
from keras.preprocessing.image import ImageDataGenerator
train_datagen=ImageDataGenerator(rescale=1./255,
                                 shear_range=0.2,
                                 zoom_range=0.2,
                                 horizontal_flip=True)
test_datagen=ImageDataGenerator(rescale=1./255)
```

Loading our data and performing data augmentation

```python
x_train = train_datagen.flow_from_directory("/content/drive/MyDrive/breastcancerprediction/BreastCancerPrediction/train",target_size=(64,64),batch_size=32,
                                             class_mode='categorical')
x_test = test_datagen.flow_from_directory("/content/drive/MyDrive/breastcancerprediction/BreastCancerPrediction/test",target_size=(64,64),batch_size=32,
                                             class_mode='categorical')
```

```python
x_train.class_indices
```

```python
x_test.class_indices
```

Intializing The Model

```python
model=Sequential()
```

Convolution Layer and pooling

```python
model.add(Convolution2D(32,(3,3),input_shape=(64,64,3),activation="relu"))
```

input shape is going to be pooled feature maps for the previous convolution

```python
model.add(MaxPooling2D(pool_size=(2,2)))
```

Flattening The Layers

```python
model.add(Flatten())
```

Adding Fully Connected Layers

```python
model.add(Dense(units=40,activation="relu",kernel_initializer="random_uniform"))
```

```python
model.add(Dense(units=2,activation="softmax",kernel_initializer="random_uniform"))
```

Summary Of The Model

```python
model.summary()
```

Compiling The CNN Model

```python
model.compile(optimizer="adam",loss="categorical_crossentropy",metrics=["accuracy"])
```

Fit The Model

```python
tr=model.fit_generator(x_train,steps_per_epoch=20,epochs=10,validation_data=x_test,validation_steps =17)
```

Save The Model

```python
model.save("cancer.h5")
```
